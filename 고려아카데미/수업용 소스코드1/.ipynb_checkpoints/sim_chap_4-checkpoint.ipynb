```python
# 문제 2

import numpy as np
plt.style.use('ggplot')

#샘플 데이터 생성
np.random.seed(2)
x = np.arange(1,201)

# 총 200 개의 데이터를 만드세요.
#y = 2 * x * ##############   왼쪽 # 부분에 입력 창 만들어 주세요.

# 위 정답 코드
y = 2 * x * np.random.rand(200)

#산포도 그래프 그리기
fig = plt.figure( )
ax = fig.add_subplot(111)
ax.scatter(x,y)
plt.show()
```
